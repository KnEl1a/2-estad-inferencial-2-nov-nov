```python
from scipy.stats import norm

def caso1_unilateral(media_muestral, mu_0, sigma, n, alpha, lado=">"):
    """ caso1_unilateral(media_muestral=83, mu_0=80, sigma=8, n=25, alpha=0.05, lado=">") # > significa derecho
    
    Realiza un contraste de hipótesis unilateral.
    
    Parámetros:
    - media_muestral: media de la muestra (X̄).
    - mu_0: media de la población bajo H0 (μ0).
    - sigma: desviación estándar poblacional.
    - n: tamaño de la muestra.
    - alpha: nivel de significancia.
    - lado: "derecho" para H_A: μ > μ0, "izquierdo" para H_A: μ < μ0.
    """
    # Calculamos el estadístico Z
    z = (media_muestral - mu_0) / (sigma / (n ** 0.5))

    # print(f"z = ")
    
    # Determinamos el valor crítico según el lado de la prueba
    if lado == ">":
        z_critico = norm.ppf(1 - alpha)  # H_A: μ > μ0
        p_valor = 1 - norm.cdf(z)
        rechazar = z > z_critico
    elif lado == "<":
        z_critico = norm.ppf(alpha)  # H_A: μ < μ0
        p_valor = norm.cdf(z)
        rechazar = z < z_critico
    
    if lado == ">": 
        print("prueba derecha (positivos)") 
    else: print("prueba izquierda (negativos)")
    print(f"Estadístico Z = {z:.4f}")
    print(f"Valor crítico Z_{alpha} = {z_critico:.4f}")
    print(f"¿ Z {lado} Z0 ? o sea {z:.4f} {lado} {z_critico:.4f} ? ")
    print(" \n -- Tenemos evidencia suficiente? ")
    print(f"dado nuestro estadistico Z = {z:.4f} nuestro: P-valor = {p_valor:.4f}")
    print("¿Rechazamos H0?", "Sí" if rechazar else "No")

# Ejemplo de prueba unilateral
caso1_unilateral(media_muestral=83, mu_0=80, sigma=8, n=25, alpha=0.05, lado=">") # > significa derecho
```

```python
# CASO 1
def caso1_bilateral(media_muestral, mu_0, sigma, n, alpha):
    """ caso1_bilateral(media_muestral=1.95, mu_0=2, sigma=0.06, n = 9, alpha=0.05)
    
    Realiza un contraste de hipótesis bilateral.
    
    Parámetros:
    - media_muestral: media de la muestra (X̄).
    - mu_0: media de la población bajo H0 (μ0).
    - sigma: desviación estándar poblacional.
    - n: tamaño de la muestra.
    - alpha: nivel de significancia.
    """
    # Calculamos el estadístico Z
    z = (media_muestral - mu_0) / (sigma / (n ** 0.5))
    
    # Valor crítico para prueba bilateral
    z_critico = norm.ppf(1 - alpha / 2)
    p_valor = 2 * (1 - norm.cdf(abs(z)))  # Se multiplica por 2 para obtener el p-valor bilateral
    
    rechazar = abs(z) > z_critico
    
    print("prueba bilateral\n") 
    print(f"Estadístico Z = {z:.4f}")
    print(f"Valor crítico Z_{alpha/2} = ±{z_critico:.4f}")
    print(f"¿?: o sea {z:.4f} esta entre -{z_critico:.4f} < Z < {z_critico:.4f}? ")
    print(f"P-valor = {p_valor:.4f}")
    print("¿Rechazamos H0?", "Sí" if rechazar else "No")

# Ejemplo de prueba bilateral
# caso1_bilateral(media_muestral=83, mu_0=80, sigma=8, n=25, alpha=0.05)
caso1_bilateral(media_muestral=1.95, mu_0=2, sigma=0.06, n = 9, alpha=0.05)
```

```python
from scipy.stats import t

def caso2_unilateral(media_muestral, mu_0, s, n, alpha, lado="<"):
    """
    Realiza un contraste de hipótesis unilateral para la media poblacional con varianza desconocida.
    
    Parámetros:
    - media_muestral: media de la muestra (X̄).
    - mu_0: media de la población bajo H0 (μ0).
    - s: desviación estándar muestral.
    - n: tamaño de la muestra.
    - alpha: nivel de significancia.
    - lado: "<" para prueba a la izquierda, ">" para prueba a la derecha.
    """
    # Calculamos el estadístico t
    t_estadistico = (media_muestral - mu_0) / (s / (n ** 0.5))
    df = n - 1  # grados de libertad

    # Valor crítico dependiendo del lado
    if lado == "<":
        t_critico = t.ppf(alpha, df)
        p_valor = t.cdf(t_estadistico, df)
    else:
        t_critico = t.ppf(1 - alpha, df)
        p_valor = 1 - t.cdf(t_estadistico, df)
    
    rechazar = (t_estadistico < t_critico) if lado == "<" else (t_estadistico > t_critico)

    # Imprimir resultados
    print(" ")
    if lado=="<": 
        print("prueba a la izquierda (negativos)") 
    else:
        print("prueba a la derecha (positivos)")
        
    print(" ")
    print("Prueba unilateral\n")
    print(f"Estadístico t = {t_estadistico:.4f}")
    print(f"Valor crítico t_{alpha} = {t_critico:.4f}")
    print("\n¿estadistico t < valor critico t_{alpha}?\n")
    
    print(f"P-valor = {p_valor:.4f}")
    print("¿Rechazamos H0?", "Sí" if rechazar else "No")

# mu >= 250
# mu < 250

# Ejemplo de uso
caso2_unilateral(media_muestral=247.5, mu_0=250, s=5.24, n=15, alpha=0.05, lado="<")
```

```python
def caso2_bilateral(media_muestral, mu_0, s, n, alpha):
    """
    Realiza un contraste de hipótesis bilateral para la media poblacional con varianza desconocida.
    
    Parámetros:
    - media_muestral: media de la muestra (X̄).
    - mu_0: media de la población bajo H0 (μ0).
    - s: desviación estándar muestral.
    - n: tamaño de la muestra.
    - alpha: nivel de significancia.
    """
    # Calculamos el estadístico t
    t_estadistico = (media_muestral - mu_0) / (s / (n ** 0.5))
    df = n - 1  # grados de libertad
    
    # Valor crítico para prueba bilateral
    t_critico = t.ppf(1 - alpha / 2, df)
    p_valor = 2 * (1 - t.cdf(abs(t_estadistico), df))

    rechazar = abs(t_estadistico) > t_critico

    # Imprimir resultados
    print("Prueba bilateral\n")
    print("alpha/2 = ", alpha/2)
    print(f"Estadístico t = {t_estadistico:.4f}")
    print(f"Valor crítico t_{alpha/2} = ±{t_critico:.4f}")
    print(f" \n pregunta : ¿ -{t_critico:.4f} < Estadistico t < {t_critico:.4f}? \n ")
    print(f" P-valor [2P(...)]= {p_valor:.4f} ")
    print(" \n pregunta: '¿Pvalor < alpha? si es asi entonces tenemos evidencia mas que suficiente para rechazar la H0\n ")
    
    print(" ¿Rechazamos H0?", "Sí" if rechazar else "No")

# Ejemplo de uso
caso2_bilateral(media_muestral=247.5, mu_0=250, s=4.94, n=15, alpha=0.05)
```

# definicion del p valor



En efecto, en este contexto, \(\chi^2\) (chi-cuadrado) es el **estadístico de prueba**. Este estadístico es una medida que cuantifica qué tan lejos está la varianza observada (o varianza muestral) de la varianza poblacional hipotética (en \(H_0\)) en función de la escala de la distribución \(\chi^2\). Al calcular \(\chi^2\), estamos evaluando si la varianza muestral observada es suficientemente extrema (es decir, inusual) para rechazar la hipótesis nula.

### Definición General del p-valor

El **p-valor** es la probabilidad de observar un valor del estadístico de prueba tan extremo (o más extremo) como el valor calculado, **asumiendo que la hipótesis nula es cierta**. En otras palabras, es la probabilidad de que, bajo \(H_0\), se obtenga un resultado tan inusual o extremo como el observado en la muestra. 

Para generalizar:

- Si el p-valor es menor que el nivel de significación \(\alpha\), entonces el resultado es lo suficientemente raro bajo \(H_0\), y se rechaza \(H_0\).
- Si el p-valor es mayor que \(\alpha\), no hay suficiente evidencia para rechazar \(H_0\).

En este caso específico, el estadístico \(\chi^2\) nos ayuda a determinar si la varianza observada en la muestra es significativamente diferente de la varianza poblacional propuesta en \(H_0\). El p-valor se interpreta como la probabilidad de obtener un \(\chi^2\) igual o mayor al observado, dado que \(H_0\) es verdadera.

```python
import numpy as np
cbv = np.array([6.1 , 9.2,  11.5,  8.6,  12.1,  3.9,  8.4,  10.1,  9.4,  8.9])

mean = np.mean(cbv)
s = np.std(cbv, ddof = 1)
print("s: ", s)
print("mean : ", mean)
```
